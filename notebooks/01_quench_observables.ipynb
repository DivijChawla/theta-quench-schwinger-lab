# Quench observables

Compute energy density, staggered mass proxy, Loschmidt rate function, and entanglement entropy.

```python
import matplotlib.pyplot as plt
from tqm.config import load_config
from tqm.pipeline import run_single_quench

cfg = load_config('../configs/default.yaml')
run = run_single_quench(cfg, theta1=2.0)

fig, ax = plt.subplots(2,2, figsize=(10,6), sharex=True)
ax[0,0].plot(run.times, run.energy_density)
ax[0,0].set_title('Energy density')
ax[0,1].plot(run.times, run.staggered_mass)
ax[0,1].set_title('Staggered mass proxy')
ax[1,0].plot(run.times, run.loschmidt_rate)
ax[1,0].set_title('Loschmidt rate')
ax[1,1].plot(run.times, run.entanglement_entropy)
ax[1,1].set_title('Bipartite entropy')
for a in ax.ravel():
    a.grid(alpha=0.2)
fig.tight_layout()
```
