# Magic vs transition diagnostics

Overlay $M_2(t)$ with the Loschmidt rate and inspect sweep heatmaps.

```python
import numpy as np
import matplotlib.pyplot as plt
from tqm.config import load_config
from tqm.pipeline import run_single_quench

cfg = load_config('../configs/default.yaml')
runs = [run_single_quench(cfg, th) for th in cfg.model.theta1_values]

fig, ax = plt.subplots(figsize=(8,4))
for r in runs:
    ax.plot(r.times, r.loschmidt_rate, alpha=0.45, lw=1.3)
ax.set_title('Loschmidt rates across quenches')
ax.set_xlabel('Time')
ax.set_ylabel('lambda(t)')
ax.grid(alpha=0.2)

overlay = runs[-1]
if 2.0 in overlay.magic:
    plt.figure(figsize=(8,4))
    plt.plot(overlay.times, overlay.loschmidt_rate, label='lambda(t)')
    plt.plot(overlay.times, overlay.magic[2.0], label='M2(t)')
    plt.legend(frameon=False)
    plt.grid(alpha=0.2)
    plt.title('Overlay for strongest quench')
    plt.xlabel('Time')
```
