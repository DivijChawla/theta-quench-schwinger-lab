# Validate small-N Schwinger quench implementation

Checks: Hermiticity, dense-vs-Krylov agreement, norm conservation, and magic sanity states.

```python
from tqm.config import load_config
from tqm.pipeline import run_single_quench, run_magic_sanity

cfg = load_config('../configs/default.yaml')
cfg.model.n_sites = 6
cfg.evolution.n_steps = 41
run = run_single_quench(cfg, theta1=1.2)
print('Hermitian H0/H1:', run.hermitian_h0, run.hermitian_h1)
print('Norm drift:', run.norm_drift)
print('Dense-Krylov error:', run.dense_krylov_error)
print('Magic sanity:', run_magic_sanity(cfg))
```
